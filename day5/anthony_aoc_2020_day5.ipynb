### Question 5.1

```python
import numpy as np
import math
import re
with open('seats.txt') as f:
    seats = [line.rstrip() for line in f]
```

```python
def rowfinder(front,back,position):
    if position =='F':
        new_front = front
        new_back = back - (((back-front)+1)/2)
    if position =='B':
        new_front = front + (((back-front)+1)/2)
        new_back = back
        
    return(int(new_front), int(new_back))
```

```python
def seatfinder(left,right,position):
    if position =='L':
        new_left = left
        new_right = right - (((right-left)+1)/2)
    if position == 'R':
        new_left = left + (((right-left)+1)/2)
        new_right = right
    return(int(new_left),int(new_right))
```

```python
correct_row = []
correct_col = []
for j in range(len(seats)):
    front = 0
    back = 127
    left = 0
    right = 7
    nfront,nback = rowfinder(front,back,seats[j][0])
    nleft,nright = seatfinder(left,right,seats[j][7])
    for i in range(1,len(seats[0])-3):
        nfront,nback = rowfinder(nfront,nback,seats[j][i])
    for m in range(8,len(seats[0])):
        nleft,nright = seatfinder(nleft,nright,seats[j][m])
    if nfront==nback:
        correct_row.append(nfront)
    if nleft==nright:
        correct_col.append(nleft)
seat_id = []
for n in range(len(correct_row)):
    seat_id.append((correct_row[n] * 8) + correct_col[n])
print(max(seat_id))
```
